```python
alphabet = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j", "k", "l", "m", "n", "o", "p", "q", "r", "s", "t", "u", "v", "w", "x", "y", "z"]
#Decode messages with a given offset
def decypher_caesar(message, offset):
    dec_message = []
    for i in range(len(message)):
        if message[i] not in alphabet:
                dec_message.append(message[i])
        else:
            for j in range(len(alphabet)):
                if alphabet[j] == message[i]:
                    if (offset + j) < len(alphabet):
                        dec_message.append(alphabet[j+offset])
                    elif (offset + j) >= len(alphabet):
                        dec_message.append(alphabet[j+offset-len(alphabet)])
    return ''.join(dec_message)
message_1 = "xuo jxuhu! jxyi yi qd unqcfbu ev q squiqh syfxuh. muhu oek qrbu je tusetu yj? y xefu ie! iudt cu q cuiiqwu rqsa myjx jxu iqcu evviuj!"
print(decypher_caesar(message_1, 10))
#Encode Messages with a given offset
def cypher_message(message, offset):
    c_message = []
    for i in range(len(message)):
        if message[i] not in alphabet:
                c_message.append(message[i])
        else:
            for j in range(len(alphabet)):
                if alphabet[j] == message[i]:
                    if (offset - j) < len(alphabet):
                        c_message.append(alphabet[j-offset])
                    elif (offset - j) >= len(alphabet):
                        c_message.append(alphabet[j-offset+len(alphabet)])
    return ''.join(c_message)
my_message = "hey there! this is an example of a caesar cipher. were you able to decode it? i hope so! send me a message back with the same offset!"
print(cypher_message(my_message, 10))
print(decypher_caesar("jxu evviuj veh jxu iusedt cuiiqwu yi vekhjuud.",10))
print(decypher_caesar("bqdradyuzs ygxfubxq omqemd oubtqde fa oapq kagd yqeemsqe ue qhqz yadq eqogdq!",14))
mess_4 = "vhfinmxkl atox kxgwxkxw tee hy maxlx hew vbiaxkl hulhexmx. px'ee atox mh kxteer lmxi ni hnk ztfx by px ptgm mh dxxi hnk fxlltzxl ltyx."
#If given a Ceaser cypher without the offset
def unkown_offset(message):
  for i in range(len(alphabet)):
    print(decypher_caesar(message,i))
    print(i)
unkown_offset(mess_4)
#Vigen√®re Cipher encode and decode 
def vigen_cipher(message,keyword):
  encode_vig = []
  key_mess = []
  index = 0
  for char in message:
    if char not in alphabet:
      key_mess.append(char)
    else:
      if index == len(keyword):
        index = 0
      key_mess.append(keyword[index])
      index += 1    
  print(''.join(key_mess))
  for i in range(len(key_mess)):
    if key_mess[i] not in alphabet:
      encode_vig.append(key_mess[i])
    else:
      encode_vig.append(alphabet[alphabet.index(message[i]) - alphabet.index(key_mess[i])])
  print(''.join(encode_vig))

def vigen_decipher(message,keyword):
  decode_vig = []
  key_mess = []
  index = 0
  for char in message:
    if char not in alphabet:
      key_mess.append(char)
    else:
      if index == len(keyword):
        index = 0
      key_mess.append(keyword[index])
      index += 1    
  print(''.join(key_mess))
  for i in range(len(key_mess)):
    if key_mess[i] not in alphabet:
      decode_vig.append(key_mess[i])
    elif (alphabet.index(message[i]) + alphabet.index(key_mess[i])) < len(alphabet):
      decode_vig.append(alphabet[alphabet.index(message[i]) + alphabet.index(key_mess[i])])
    elif (alphabet.index(message[i]) + alphabet.index(key_mess[i])) >= len(alphabet):
      decode_vig.append(alphabet[alphabet.index(message[i]) + alphabet.index(key_mess[i])- len(alphabet)])
  print("".join(decode_vig))

vigen_cipher("barry is the spy","dog")
vigen_decipher("txm srom vkda gl lzlgzr qpdb? fepb ejac! ubr imn tapludwy mhfbz cza ruxzal wg zztylktoikqq!","friends")
```
